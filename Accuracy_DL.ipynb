```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# Simulate the true labels and predicted labels
# 0 = Cat, 1 = Dog, 2 = Lion, 3 = Tiger
true_labels = np.array([0]*3000 + [1]*3000 + [2]*20 + [3]*180)
predicted_labels = np.array([0]*2900 + [1]*100 + [1]*2900 + [0]*100 + [2]*15 + [3]*5 + [3]*170 + [2]*10)

# Calculate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cat', 'Dog', 'Lion', 'Tiger'],
            yticklabels=['Cat', 'Dog', 'Lion', 'Tiger'])
plt.title("Confusion Matrix")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Calculate accuracy for each class
class_accuracies = cm.diagonal() / cm.sum(axis=1) * 100

# Print accuracy for each class
class_names = ['Cat', 'Dog', 'Lion', 'Tiger']
for i, class_name in enumerate(class_names):
    print(f"Accuracy for {class_name}: {class_accuracies[i]:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix, precision_recall_curve, roc_curve, auc, classification_report
from sklearn.preprocessing import label_binarize
from sklearn.metrics import precision_recall_curve, roc_curve, auc

# Simulate the true labels and predicted labels
# 0 = Cat, 1 = Dog, 2 = Lion, 3 = Tiger
true_labels = np.array([0]*3000 + [1]*3000 + [2]*20 + [3]*180)
predicted_labels = np.array([0]*2900 + [1]*100 + [1]*2900 + [0]*100 + [2]*15 + [3]*5 + [3]*170 + [2]*10)

# Calculate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cat', 'Dog', 'Lion', 'Tiger'],
            yticklabels=['Cat', 'Dog', 'Lion', 'Tiger'])
plt.title("Confusion Matrix")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Calculate accuracy for each class
class_accuracies = cm.diagonal() / cm.sum(axis=1) * 100

# Print accuracy for each class
class_names = ['Cat', 'Dog', 'Lion', 'Tiger']
for i, class_name in enumerate(class_names):
    print(f"Accuracy for {class_name}: {class_accuracies[i]:.2f}%")

# Plot Precision-Recall Curve for each class
plt.figure(figsize=(8, 6))
for i in range(4):
    precision, recall, _ = precision_recall_curve(true_labels == i, predicted_labels == i)
    plt.plot(recall, precision, label=f'{class_names[i]}')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc='best')
plt.show()

# Plot ROC Curve for each class
plt.figure(figsize=(8, 6))
for i in range(4):
    fpr, tpr, _ = roc_curve(true_labels == i, predicted_labels == i)
    plt.plot(fpr, tpr, label=f'{class_names[i]} (AUC = {auc(fpr, tpr):.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='best')
plt.show()

# Plot Class Distribution Bar Chart
class_counts = np.bincount(true_labels, minlength=4)
plt.figure(figsize=(8, 6))
sns.barplot(x=class_names, y=class_counts)
plt.title('Class Distribution')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.show()

# Plot Prediction Probability Distribution
predicted_probs = np.array([np.mean(predicted_labels == i) for i in range(4)])
plt.figure(figsize=(8, 6))
sns.barplot(x=class_names, y=predicted_probs)
plt.title('Prediction Probability Distribution')
plt.xlabel('Class')
plt.ylabel('Mean Probability')
plt.show()

# Misclassification Matrix (without diagonal)
misclassification_matrix = cm.copy()
np.fill_diagonal(misclassification_matrix, 0)

plt.figure(figsize=(8, 6))
sns.heatmap(misclassification_matrix, annot=True, fmt='d', cmap='Reds',
            xticklabels=['Cat', 'Dog', 'Lion', 'Tiger'],
            yticklabels=['Cat', 'Dog', 'Lion', 'Tiger'])
plt.title("Misclassification Matrix")
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Mill MLP Model Results
mlp_loss = [0.5098, 0.4886, 0.5098, 0.4890, 0.4968, 0.5073, 0.5113, 0.5086, 0.4971, 0.5041]
mlp_val_loss = [0.5000] * 10
mlp_accuracy = [None] * 10  # Since no accuracy data was provided
mlp_val_accuracy = [None] * 10

# Single Convolutional Layer Model Results
conv1_loss = [1.6886, 1.3055, 1.1901, 1.0040, 0.9561, 0.8720, 0.7632, 0.7027, 0.6687, 0.5917]
conv1_val_loss = [2.1707, 1.6558, 1.0894, 0.8220, 0.6154, 0.5749, 0.5100, 0.4714, 0.4647, 0.4436]
conv1_accuracy = [0.3900, 0.5337, 0.5493, 0.5921, 0.6231, 0.6406, 0.6884, 0.7280, 0.7459, 0.7474]
conv1_val_accuracy = [0.2944, 0.3333, 0.5000, 0.6167, 0.7611, 0.7556, 0.7667, 0.7944, 0.8056, 0.8056]

# Bigger Convolutional Model Results
conv2_loss = [13.7353, -0.8413, -7.5324, -12.6320, -18.9785, -24.2836, -30.4059, -37.6381, -41.8547, -48.2857]
conv2_val_loss = [-0.3374, -6.5604, -12.3020, -17.9502, -23.6143, -29.3598, -35.6149, -42.1223, -48.9316, -55.7176]
conv2_accuracy = [None] * 10  # Since no accuracy data was provided
conv2_val_accuracy = [None] * 10

# Huge Model Results
huge_loss = [11.7663, 5.8371, 3.0217, 2.2945, 1.8178, 1.4166, 1.3035, 1.2487, 1.0156, 0.8894]
huge_val_loss = [6.5424, 2.9706, 2.3948, 1.9572, 1.7838, 1.6607, 1.5428, 1.4897, 1.3896, 1.3144]
huge_accuracy = [0.3320, 0.3735, 0.4231, 0.5131, 0.5679, 0.6101, 0.6183, 0.6407, 0.6735, 0.7173]
huge_val_accuracy = [0.3667, 0.4722, 0.4722, 0.5278, 0.5778, 0.5389, 0.5667, 0.6111, 0.6333, 0.6167]

# Function to plot training and validation metrics
def plot_metrics(loss, val_loss, accuracy, val_accuracy, title):
    epochs = range(1, len(loss) + 1)

    plt.figure(figsize=(14, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, loss, 'bo-', label='Training loss')
    plt.plot(epochs, val_loss, 'ro-', label='Validation loss')
    plt.title(f'{title} - Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    if accuracy[0] is not None:
        plt.plot(epochs, accuracy, 'bo-', label='Training accuracy')
        plt.plot(epochs, val_accuracy, 'ro-', label='Validation accuracy')
        plt.title(f'{title} - Accuracy')
        plt.xlabel('Epochs')
        plt.ylabel('Accuracy')
        plt.legend()
    else:
        plt.title(f'{title} - No Accuracy Data Available')

    plt.show()

# Plot the metrics for each model
plot_metrics(mlp_loss, mlp_val_loss, mlp_accuracy, mlp_val_accuracy, "Mill MLP")
plot_metrics(conv1_loss, conv1_val_loss, conv1_accuracy, conv1_val_accuracy, "Single Convolutional Layer")
plot_metrics(conv2_loss, conv2_val_loss, conv2_accuracy, conv2_val_accuracy, "Bigger Convolutional Model")
plot_metrics(huge_loss, huge_val_loss, huge_accuracy, huge_val_accuracy, "Huge Model")

# Print comparison of final validation accuracy
print("Validation Accuracy Comparison:")
print(f"Mill MLP: {mlp_val_accuracy[-1] if mlp_val_accuracy[0] is not None else 'N/A'}")
print(f"Single Convolutional Layer: {conv1_val_accuracy[-1]}")
print(f"Bigger Convolutional Model: {conv2_val_accuracy[-1] if conv2_val_accuracy[0] is not None else 'N/A'}")
print(f"Huge Model: {huge_val_accuracy[-1]}")
```
